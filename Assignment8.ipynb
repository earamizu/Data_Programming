```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
%matplotlib inline

#Importing data set
sns.set_style('whitegrid')
titanic = sns.load_dataset('titanic')
# titanic.head()

#Code was already partially presented on Exercise
#Joint plot showing correlation coeficient and p-value
sns.jointplot(x='fare',y='age',data=titanic, stat_func=stats.pearsonr,\
              marginal_kws={'hist_kws': {'histtype': 'step', 'fill': 'b', 'alpha': 0.55}})
```

```python
#Code was already partially presented on Exercise
#Histogram from fare variable.
sns.distplot(titanic['fare'],bins=30,kde=False,color='RED', hist_kws={'histtype': 'step', 'fill': 'r',\
                                                                     'alpha': 0.55})
```

```python
#Code was already presented on Exercise
#Box plot showing age by class
sns.boxplot(x='class', y='age', data=titanic, palette='rainbow')
```

```python
#Code was already presented on Exercise
#Swarm plot from class by age
sns.swarmplot(x='class', y='age', data=titanic, palette='rainbow')
```

```python
#Code was already presented on Exercise
#Count plot by sex
sns.countplot(x='sex', data=titanic, palette=sns.xkcd_palette(['windows blue', 'faded green']))
```

```python
#Correlation Matrix
#Plot heat map of correlation matrix
sns.heatmap(titanic.corr(), center=0, vmin=-1, vmax=1, cmap="coolwarm")
```

```python
#Faceting histogram of ages by sex
g = sns.FacetGrid(titanic, col='sex')
g = g.map(sns.distplot, 'age', kde=False, bins=10, color="#4472C4", hist_kws=dict(alpha=1, edgecolor="k", linewidth=1))
```
