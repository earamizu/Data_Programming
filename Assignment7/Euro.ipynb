```python
import pandas as pd
import urllib.request
import io

url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"

try:
    fhand = urllib.request.urlopen(url)
except:
    print("URL invalid")
    exit()
    
euro12 = pd.read_csv(io.StringIO(fhand.read().decode('utf-8')))

#Step 4
print(f"Goals Column\n{euro12['Goals']}\n")
      
#Step 5
print(f"Total of teams is {len(euro12)}\n")

#Step 6
print(f"Total of columns from dataset is {len(euro12.columns)}\n")
# euro12.columns.values

#Step 7
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']].copy()

#Step 8
sorted_by_cards = discipline.sort_values(['Red Cards', 'Yellow Cards'])
print(f"Teams sorted by Red and Yellow cards\n{sorted_by_cards}\n")
      
#Step 9
avg_yellow_cards = sorted_by_cards['Yellow Cards'].mean()
print(f"Average of yellow cards per team is {avg_yellow_cards}\n")

#Step 10
print(f"Teams that scored more than 6 goals\n{euro12[euro12['Goals'] > 6][['Team', 'Goals']]}\n")
      
#Step 11
print(f"Teams starting with \'G\'\n{euro12[euro12['Team'].astype(str).str[0] == 'G']['Team']}\n")

#Step 12
print(f"First 7 columns\n{euro12.iloc[:,0:7]}\n")

#Step 13
t_col = len(euro12.columns)
print(f"All columns but last 3\n{euro12.iloc[:,:t_col-3]}")

#Step 14
euro_indexed = euro12.set_index(['Team'])
print(f"Shooting accuracy from England, Italy and Russia\n{euro_indexed.loc[['England', 'Italy', 'Russia']]['Shooting Accuracy']}\n")
```
