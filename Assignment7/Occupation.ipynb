```python
import pandas as pd
import io
import urllib.request

url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
try:
    fhand = urllib.request.urlopen(url)
except:
    print('URL invalid')
    exit()
    
users = pd.read_csv(io.StringIO(fhand.read().decode('utf-8')), sep='|')

#Step 4
avg_age_by_occ = pd.DataFrame(users.groupby(['occupation'])['age'].mean()).round({'age': 2})
print(f"Mean age by occupation\n{avg_age_by_occ}\n")

#Step 5
aux = users.groupby(['occupation', 'gender']).count()
aux = pd.pivot_table(users, values='user_id', index=['occupation',],
                    columns=['gender'], aggfunc='count', fill_value=0)
aux['male_ratio'] = round(aux['M']/(aux['M']+aux['F']), 2)
male_ratio = pd.DataFrame(aux['male_ratio'].copy()).sort_values('male_ratio', ascending=False)
print(f"Male ratio per occupation\n{male_ratio}\n")

#Step 6
max_min_by_occ = pd.pivot_table(users, values=['age'], index=['occupation'],
                                aggfunc={'age': [min, max]}, fill_value=0)

print(f"Maximum and Minimum ages per occupation\n{max_min_by_occ}\n")

#Step 7
avg_age_by_occ_gender = pd.pivot_table(users, values=['age'], index=['occupation'], columns=['gender'],
                                      aggfunc='mean', fill_value=0).round(2)
print(f"Mean age by occupation and gender\n{avg_age_by_occ_gender}\n")

#Step 8
aux = pd.pivot_table(users, values='user_id', index=['occupation'], columns=['gender'],
                     aggfunc='count', fill_value=0)
aux['%F'] = round(aux['F']/(aux['F']+aux['M']), 2)
aux['%M'] = round(aux['M']/(aux['F']+aux['M']), 2)
percent_by_gender = aux[['%F', '%M']]
print(f"Percentage of women and men\n{percent_by_gender}")
```
